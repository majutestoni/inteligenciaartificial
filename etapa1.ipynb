```python

n = 10
matriz = [[0 for _ in range(n)] for _ in range(n)]
direita = False
cima  = False
baixo = False
esquerda = False
i = 7
j = 7

def vaiDireita():
    global j
    j += 1

def vaiEsquerda():
    global j
    j -= 1

def vaiBaixo():
    global i
    i -= 1

def vaiCima():
    global i
    i+=1

def podeDireita():
    global j,n,direita
    return j < n -1 and not direita

def podeBaixo():
    global i,n,baixo
    return i < n -1 and not baixo

def podeEsquerda():
    global i,n,esquerda,j
    return i == n -1 and not esquerda and j != 0

def podeCima():
    global i,cima
    return i > 0 and not cima

while (not direita or not esquerda or not baixo or not cima):
      matriz[i][j] = 1

      if podeDireita():
        vaiDireita()
      elif podeBaixo() :
        direita = True
        vaiCima()
      elif podeEsquerda():
        baixo = True
        vaiEsquerda()
      elif podeCima():
        esquerda = True
        vaiBaixo()
      else:
          cima = True


for linha in matriz:
    print(linha)
```
